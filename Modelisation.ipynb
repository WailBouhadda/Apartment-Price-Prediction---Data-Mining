```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Prix']
X = df.loc[:, df.columns != 'Prix']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=45, test_size=0.2)
```

# 1 - Linear Regression

### a - Create Model

```python
from sklearn.linear_model import LinearRegression
```

```python
LinReg = LinearRegression()
```

```python
LinReg.fit(X_train, y_train)
```

### b - Test Model

```python
import matplotlib.pyplot as plt
```

```python
y_pred = LinReg.predict(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
n = pca.fit_transform(X_test[['Etage','Surface habitable', 'NB Salons', 'NB Chambres', 'NB Salles de Bain', 'Longitude', 'Latitude']])
```

```python
plt.scatter(n, y_test, color='red', linewidth=1, label='Actual Data')
plt.scatter(n, y_pred, color='blue', linewidth=1, label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
```

```python
import numpy as np
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error: {rmse}')
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2}')
```
